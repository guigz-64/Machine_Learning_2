# Libraries Importation

```python
import numpy as np
```

# Declaration of Variables

```python
a = np.array([[1,0,0,0],[0.2,0.3,0.1,0.4],[0.2,0.5,0.2,0.1],[0.7,0.1,0.1,0.1]]) # Transition Probabilities
b = np.array([[1,0,0,0,0],[0,0.3,0.4,0.1,0.2],[0,0.1,0.1,0.7,0.1],[0,0.5,0.2,0.1,0.2]]) # Emission Probabilities

N = 3
T = 4

sequence_k = np.array([1,3,2,0]) #Sequence
```

# 1. Evaluation Problem

```python
def evaluation(N:int, T:int, a, b, sequence_k):
    alpha = np.zeros(shape=(N+1,T+1))
    alpha[:,0] = [0,1,0,0]

    for t in range(1, T+1):
        for j in range(0, N+1):
            alpha[j,t] = sum( alpha[i,t-1]*a[i,j]*b[j,sequence_k[t-1]] for i in range(1,N+1) )
    
    print(alpha)
    return alpha[0, T]
```

```python
P_X_given_M = evaluation(N,T,a,b,sequence_k)
print(f"\nP(X | M) = {P_X_given_M}")
```

# 2. Decoding Problem

```python
def decoding(N:int, T:int, a, b, sequence_k):
    alpha = np.zeros(shape=(N+1,T+1))
    alpha[:,0] = [0,1,0,0]
    path = [f'Z{sequence_k[0]}']

    for t in range(1, T+1):
        for j in range(0, N+1):
            alpha[j,t] = sum( alpha[i,t-1]*a[i,j]*b[j,sequence_k[t-1]] for i in range(1,N+1) )
        path.append(f'Z{np.argmax(alpha[:,t])}')

    return path
```

```python
path = decoding(N,T,a,b,sequence_k)
print(path)
```
